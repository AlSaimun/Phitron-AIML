```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

## 2D line plot

```python
hours = [2 , 3 , 4, 1 , 5 , 7 ,3] 
days = [1 ,2 , 3 ,4 ,5 ,6 ,7] 

plt.plot(days,hours)
```

```python
plt.xlabel('Days')
plt.ylabel('Study Hours') 
plt.title('Trend of a student study time over a week')
plt.plot(days, hours)
plt.show()
```

## 2D line plot from pandas data

```python
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

df = pd.read_csv('enrollment_data.csv')
plt.plot(df['Year'], df['Programming'])
```

```python
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

df = pd.read_csv('enrollment_data.csv')
plt.plot(df['Year'], df['Programming'])
plt.plot(df['Year'], df['Digital Marketing'])
```

```python
plt.plot(df['Year'], df['Programming'], label='Programming')
plt.plot(df['Year'], df['Digital Marketing'], label='Digital Marketing')
plt.legend()
```

```python
plt.plot(df['Year'], df['Programming'], label='Programming')
plt.plot(df['Year'], df['Digital Marketing'], label='Digital Marketing')
plt.legend(loc='upper center')
```

```python
plt.plot(df['Year'], df['Programming'], label='Programming')
plt.plot(df['Year'], df['Digital Marketing'], label='Digital Marketing')
plt.legend(loc='best')
```

```python
# adding some visualization 

plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

df = pd.read_csv('enrollment_data.csv')
plt.plot(df['Year'], df['Programming'], color='red', linewidth=3, linestyle='dashed', marker='o', markersize=7)
plt.plot(df['Year'], df['Digital Marketing'], color='blue', linewidth=3, linestyle='dotted', marker='o', markersize=7)
```

```python
# adding some visualization 

plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'] ,df['Programming'],label = 'Programming', color ='#11b86f',linewidth=3,linestyle='dashed',marker='o',markersize=7)
plt.plot(df['Year'] ,df['Digital Marketing'], label= 'Digital Marketing', color='black',linewidth=3,linestyle='dashed',marker='o',markersize=7)

plt.legend()
plt.grid()

plt.show()
```

```python
students = pd.read_csv('student_IQdata.csv')

students
```

```python
#Relation between Study Hours and IQ Score - postive correlation

plt.xlabel('Study Hours')
plt.ylabel('IQ Score')
plt.title('Relation between Study Hours and IQ Score') 

plt.scatter(students['Study_Hour'], students['IQ_Score'])
```

```python
#Relation between Shoe size and IQ Score - > no correlation

plt.xlabel('Shoe Size')
plt.ylabel('IQ Score')
plt.title('Relation between Shoe Size and IQ Score') 

plt.scatter(students['Shoe_Size'] , students['IQ_Score']) 
```

```python
#Relation between Chilling hour and IQ Score - > negative correlation
plt.xlabel('Chilling Time')
plt.ylabel('IQ Score')
plt.title('Relation between Chilling hours and IQ Score')

plt.scatter(students['Chilling_Hours'] , students['IQ_Score'],marker='o',color='green') 
```

```python
plt.hist(students['IQ_Score'],color='green',edgecolor='black')
plt.xlabel('IQ Score')
plt.ylabel('Frequency of Students')
```
