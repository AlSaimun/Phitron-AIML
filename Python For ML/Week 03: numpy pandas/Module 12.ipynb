## Transform df to inbuilt datastructure

```python
my_tuple = (('Alice', 25), ('Bob', 30), ('Charlie', 28))

tuple_df = pd.DataFrame(my_tuple,columns=['Name','Age'])

tuple_df
```

```python
my_dict = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 29,28]
}

dict_df = pd.DataFrame(my_dict)

dict_df
```

## Sort

```python
sorted_df = df.sort_values('Data Structure Marks')
print(sorted_df.head(5))
sorted_df = df.sort_values(['Data Structure Marks', 'Python Marks'])
sorted_df.head()
```

```python
# decending 
sorted_df = df.sort_values(['Data Structure Marks', 'Python Marks'], ascending=False)
sorted_df
```

```python
# decending 
sorted_df = df.sort_values(['Data Structure Marks', 'Python Marks'], ascending=[0, 1])
sorted_df
```
