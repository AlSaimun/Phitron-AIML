```python
index_name = ["a", "b", "c", "d", "e"]
dfs = df['Name'][0:5]
dfs
```

```python
new_df = pd.Series([1, 2, 3, 4, 5], index=index_name)
new_df
```

```python
new_dfs = pd.Series(list(dfs), index=index_name)
new_dfs
```
