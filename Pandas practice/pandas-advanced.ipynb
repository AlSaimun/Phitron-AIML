## Join method

```python
data1 = pd.DataFrame({
    'Column1' : [1, 2,4, 5, 6],
    'Column2' : [4, 5, 6, 7, 8],
    'Column3' : [3, 4, 5, 6, 6]
}, index = ['a', 'b', 'c', 'd', 'e'])

data1
```

```python
data2 = pd.DataFrame({
    'Column11' : [1, 2,4, 5, 6],
    'Column22' : [4, 5, 6, 7, 8],
    'Column33' : [3, 4, 5, 6, 6]
}, index = ['a', 'b', 'h', 'i', 'j'])

data2
```

```python
data1.join(data2)
```

```python
data1.join(data2, how='right')
```

```python
data1.join(data2, how='inner')
```

```python
data1.join(data2, how='outer')
```

```python
data1.join(data2, how='cross')
```

## Apply 

### apply method এ প্যারামিটার হিসেবে একটা ফাংশন নেয়। এরপর এক এক করে ডেটা ফ্রেম থেকে যে কলামের উপর apply method চালানো হয়েছে সেটা থেকে ডেটা নিবে এবং ফাংশনে পাস করবে। এরপর ফাংশন ঐটা ডেটা উপর অপারেশন করে দিবে। 

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df.head()
```

```python
df['Taka'] = df['Fare'].apply(lambda x: x*100)
df.head()
```

```python
df['Name Len'] = df['Name'].apply(len)
df.head()
```

```python
df['Name Len'][:5]
```

```python
df['Category Of Fare'] = df['Fare'].apply(lambda x : 'Chip' if x < 10 else ('Mid' if x > 10 and x < 50 else 'High'))
df.head()
```
